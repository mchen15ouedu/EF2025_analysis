```python
import os
import numpy as np
import datetime as dt
import rasterstats as rs
import pandas as pd
import geopandas as gpd
import netCDF4 as nc
```

```python
dircG='e:/PeruProject/WRF_var/w_RUNOFF/'
dircW='e:/PeruProject/WRF_var/wrf_runoff/'
```

```python
times=pd.date_range('20150101','20200101',freq='3H', closed='left')
print(times)
```

```python
test=nc.Dataset(dircW+'201501010000.LDASOUT_DOMAIN1.nc')
print(test.variables)
```

```python
acc_g_sur=np.zeros((310,330),dtype='float')
acc_g_sub=np.zeros((310,330),dtype='float')
acc_w_sur=np.zeros((330,310),dtype='float')
acc_w_sub=np.zeros((330,310),dtype='float')
```

```python
df=[]
for t in times:
    gld_f=dircG+t.strftime('%Y%m%d%H')+'00.LDASOUT_DOMAIN1.nc'
    wrf_f=dircW+t.strftime('%Y%m%d%H')+'00.LDASOUT_DOMAIN1.nc'
    gldas=nc.Dataset(gld_f)
    wrf=nc.Dataset(wrf_f)
    print('finished reading')
    
    gldas_sur=gldas.variables['SFCRNOFF'][:]
    gldas_sub=gldas.variables['UGDRNOFF'][:]
    
    wrf_sur=wrf.variables['runoff'][:]
    wrf_sub=wrf.variables['subrunoff'][:]
    
    gldas_sur[gldas_sur<0]=np.nan
    gldas_sub[gldas_sub<0]=np.nan
    wrf_sur[wrf_sur<0]=np.nan
    wrf_sub[wrf_sub<0]=np.nan
    
    acc_g_sur=acc_g_sur+gldas_sur[0]*3
    acc_g_sub=acc_g_sub+gldas_sub[0]*3
    acc_w_sur=acc_w_sur+wrf_sur[0]*3
    acc_w_sub=acc_w_sub+wrf_sub[0]*3
    
    m=[np.nanmean(gldas_sur),np.nanmean(gldas_sub),np.nanmean(wrf_sur),np.nanmean(wrf_sub)]
    df.append(m)
    
    print('finish calculating '+t.strftime('%Y-%m-%d'))
    
    gldas.close()
    wrf.close()
```

```python
roff=pd.DataFrame(df)
roff.columns=['GLDAS_surface_runoff','GLDAS_subsurface_runoff','WRF_surface_runoff','WRF_subsurface_runoff']
print(roff)
```

```python
roff.to_csv('e:/PeruProject/3rdpaper/runoff.csv')
```

```python
from osgeo import gdal, osr
```

```python
x_g_res=(-70.808333+72.791667)/310
y_g_res=(-14.929167+16.854167)/330

x_w_res=(-70.808333+72.791667)/330
y_w_res=(-14.929167+16.854167)/310

g_geo=(-72.791667,x_g_res,0,-14.929167,0,-y_g_res)
w_geo=(-72.791667,x_w_res,0,-14.929167,0,-y_w_res)
```

```python
dirc_out='e:/PeruProject/3rdpaper/'
fname='wrf_surface_runoff.tif'
array=acc_w_sur/5

driver=gdal.GetDriverByName('GTiff')
ann_dat=driver.Create(dirc_out+fname,310,330,1,gdal.GDT_Float32,options=['COMPRESS=Deflate'])
ann_dat.SetGeoTransform(w_geo)
srs=osr.SpatialReference()
srs.ImportFromEPSG(4326)
ann_dat.SetProjection(srs.ExportToWkt())
ann_dat.GetRasterBand(1).WriteArray(array)
ann_dat.GetRasterBand(1).SetNoDataValue(-9999)
ann_dat.FlushCache()

ann_dat=None
```

```python
dirc_out='e:/PeruProject/3rdpaper/'
fname='wrf_subsurface_runoff.tif'
array=acc_w_sub/5

driver=gdal.GetDriverByName('GTiff')
ann_dat=driver.Create(dirc_out+fname,310,330,1,gdal.GDT_Float32,options=['COMPRESS=Deflate'])
ann_dat.SetGeoTransform(w_geo)
srs=osr.SpatialReference()
srs.ImportFromEPSG(4326)
ann_dat.SetProjection(srs.ExportToWkt())
ann_dat.GetRasterBand(1).WriteArray(array)
ann_dat.GetRasterBand(1).SetNoDataValue(-9999)
ann_dat.FlushCache()

ann_dat=None
```

```python
df=[]
for i in reversed(range(len(times))):
    gld_f=dircG+times[i].strftime('%Y%m%d%H')+'00.LDASOUT_DOMAIN1.nc'
    gld_p=dircG+times[i-1].strftime('%Y%m%d%H')+'00.LDASOUT_DOMAIN1.nc'
    gldas=nc.Dataset(gld_f)
    sub=nc.Dataset(gld_p)
    print('finished reading')
    
    gldas_sur=gldas.variables['SFCRNOFF'][:]
    gldas_sub=gldas.variables['UGDRNOFF'][:]
    
    s_sur=sub.variables['SFCRNOFF'][:]
    s_sub=sub.variables['UGDRNOFF'][:]
    
    gldas_sur[gldas_sur<0]=np.nan
    gldas_sub[gldas_sub<0]=np.nan
    s_sur[s_sur<0]=np.nan
    s_sub[s_sub<0]=np.nan
    
    #acc_g_sur=acc_g_sur+(gldas_sur[0]-s_sur[0])*3
    #acc_g_sub=acc_g_sub+(gldas_sub[0]-s_sub[0])*3

    m=[np.nanmean(gldas_sur-s_sur),np.nanmean(gldas_sub-s_sub)]
    df.append(m)
    
    print('finish calculating '+times[i].strftime('%Y-%m-%d %H:00'))
    
    gldas.close()
    sub.close()
```
